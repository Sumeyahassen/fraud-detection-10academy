```python
# import library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
fraud=pd.read_csv('../data/raw/Fraud_Data.csv')
ip_country=pd.read_csv('../data/raw/IpAddress_to_Country.csv')
credit=pd.read_csv('../data/raw/creditcard.csv')
print(fraud.shape,ip_country.shape,credit.shape)
```

```python
print(fraud.info())
print(credit.info())
print(ip_country.info())
```

```python
# fraud data clening
print(f'missing value {fraud.isnull().sum()}') # type: ignore
print(f'duplication data {fraud.duplicated().sum()}') # type: ignore
fraud=fraud.drop_duplicates()
```

```python
# conver timestamp
fraud['signup_time']=pd.to_datetime(fraud['signup_time'])
fraud['purchase_time']=pd.to_datetime(fraud['purchase_time'])
```

```python
# credit card cleaning
print(f'missing credite {credit.isnull().sum()}')
print(f"duplication {credit.duplicated().sum()}")
credit=credit.drop_duplicates()
credit.duplicated().sum()
```

```python
# explatori data analaysis
# fraud
print('fraud data class desterbution')
print(fraud['class'].value_counts(normalize=True))
# credit card
print("cradit card destribution")
print(credit['Class'].value_counts(normalize=True))
```

```python
# uni and bi variant
# fraud vizualization
sns.countplot(x='class',data=fraud)
plt.title="class destributon in e-commerse"
plt.show()
```

```python
# vizualize using histogram
sns.histplot(data=fraud, x='purchase_value' ,hue='class',bins=50)
print("perchase class destribution in class")
plt.show()
```

```python
# data visualise by broser,sex,source 
sns.catplot(data=fraud,x='browser',hue='class')
plt.show
sns.catplot(data=fraud,x='sex',hue='class')
plt.show
```

```python
# source visualise
sns.barplot(data=fraud,x='source',hue='class')
plt.show
```
